```python
# From the two data Inflation has increased significantly, but GDP growth has also increased.
#Test Null Average Inflation, consumer prices (annual %) & GDP Growth are Equal - 2 Sample Ind TTest
from scipy.stats import ttest_ind
```

```python
ttest_ind(philippines_ind.iloc[:,0],philippines_ind.iloc[:,2],equal_var=False)
# Since pvalue=0.16230453948428655 > 0.05 , hence fail to reject null
# The above two statements is to test null hypothesis whether 
# the mean of above two variables are equal or not considering variances are unequal
# Since, we have failed to reject null hypothesis, it help us to understand that 
# the mean of both variables is equal and is statistically significant.
```

```python
print("Inflation, consumer prices (annual %):",np.var(philippines_ind.iloc[:,2]))    #answer(f)
print("Inflation, GDP deflator (annual %):",np.var(philippines_ind.iloc[:,1]))
print("GDP growth (annual %):",np.var(philippines_ind.iloc[:,0]))
print("Population growth (annual %):",np.var(philippines_ind.iloc[:,3]))
```

```python
y=Indicatorsdf['GDP growth (annual %)']
```

```python
x=Indicatorsdf[['Population growth (annual %)','CO2 emissions (metric tons per capita)',
                 'Industry (including construction), value added (% of GDP)',
                 'Exports of goods and services (% of GDP)','Imports of goods and services (% of GDP)',
                 'Gross capital formation (% of GDP)','Inflation, consumer prices (annual %)']]
```

```python
x.corr()
```

```python
# Since, in none of the pair of variables, correlation is more than .95. 
# Thus, we can say that that no multicollinearity exists. 
```

```python
from sklearn.linear_model import LinearRegression  #answer(l)
# Multiple linear regression.
```

```python
reg=LinearRegression()
```

```python
regmodel=reg.fit(x,y)
```

```python
regmodel.score(x,y) 
# Since R square value is 0.6324769047972499. So, the independent variables
# can explain 63%of the changes in independent variable i.e. GDP growth.
```

```python
regpredict=regmodel.predict(x)
```

```python
regresidual=regmodel.predict(x)
```

```python
regresidual=y-regpredict
```

```python
np.sqrt(np.mean(regresidual**2)) 
# RMSE takes into consideration the square of error in forecasted 
# value over actual values. Thus, gives a much more accuracy.
# In this case case is very low which teels that our model is accurately
# able to predict the values with existing variables 
# but must be considered in context of R square.
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree=DecisionTreeRegressor()
```

```python
treemodel=tree.fit(x,y)
```

```python
treemodel.score(x,y)
# Since, getting a score of 1 is next to impossible in real life situation.
# Thus, we can interpret that decision tree has led to overfitting.
# So, we need to specify the depth to limit the clusters it forms.
```

```python
tree2=DecisionTreeRegressor(max_depth=4)
#Reducing the depth to 4 we get the score within the range i.e 0.6-0.95
```

```python
tree2model=tree2.fit(x,y)
```

```python
tree2model.score(x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor(n_estimators=1000)
```

```python
RFmodel=RF.fit(x,y)
```

```python
RFmodel.score(x,y)
# The score is low compared to decision tree for random forest 
# as it does not lead to overfitting. Thus, they are robust 
# since they consider various parameters as well as random subset of training data to form decision tree.
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbm=GradientBoostingRegressor(n_estimators=3000)
```

```python
gbmodel=gbm.fit(x,y)
```

```python
gbmodel.score(x,y)
# Since, the result we get from Gradient Boosting Method is 0.999999. 
# Thus, we can conveniently conclude that the method has accurately read
# the training data as it corrects the errors made by previous trees. But such
# high levels of accuracy can be a sign of overfitting.
```
